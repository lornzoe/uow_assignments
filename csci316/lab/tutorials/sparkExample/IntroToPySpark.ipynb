# DataFrame rows

```python
# Reading a new practice data set
df2PySpark = spark.read.csv('PracticeData2.csv', header = True, inferSchema = True)
df2PySpark.show(30)
df2PySpark.printSchema()
```

#### Filtering rows

```python
# Dropping or removing rows containing null values
df21PySpark = df2PySpark.na.drop()
```

```python
df2PySpark.show(30)
df21PySpark.show(30)
```

#### Notice that all rows that contain 'null' values will be dropped or removed

```python
# Dropping or removing rows containing 'null' values with condition
# using 'how = all' condition
# Note: 'how = all' means rows with all attributes values equal null will be removed.
df22PySpark = df2PySpark.na.drop(how = 'all')
df2PySpark.show(30)
df22PySpark.show(30)
```

```python
# Dropping or removing rows containing 'null' values with condition
# using 'how = any' condition
# Note: how = any means rows where any of the attributes value is null will be removed
df23PySpark = df2PySpark.na.drop(how = 'any')
df2PySpark.show(30)
df23PySpark.show(30)
```

#### Notice the result of drop(how = 'any') is similar to just drop(). Hence, how = 'any' is the default option.

```python
# Dropping or removing rows containing 'null' values with condition
# using 'thresh = some value' condition
# Thresh = 1 means rows with at least 1 or more 'NOT null' values will be retained.
df24PySpark = df2PySpark.na.drop(how = 'any', thresh = 1)
df2PySpark.show(30)
df24PySpark.show(30)
```

#### Notice from the result shown above, the row with all null values (the row between John and Kevin) is removed because the row contains less than 1 attributes with NOT null value.

```python
# Dropping or removing rows containing 'null' values with condition
# using 'thresh = some value' condition
# Thresh = 2 means rows with at least 2 or more'NOT null' values will be ratained.
df24PySpark = df2PySpark.na.drop(how = 'any', thresh = 2)
df2PySpark.show(30)
df24PySpark.show(30)
```

#### From the result shown above, the rows with less than 2 NOT null values are removed. That is, the row between John and Kevin, the row for Max, and the last three rows from df2PySpark.

```python
# Dropping or removing rows containing 'null' values with condition
# using 'thresh = some value' condition
# Thresh = 3 means rows with at least 3 or more 'NOT null' values will be ratained.
df24PySpark = df2PySpark.na.drop(how = 'any', thresh = 3)
df2PySpark.show(30)
df24PySpark.show(30)
```

```python
# Dropping or removing rows containing 'null' values with condition
# using 'subset = column name' condition
df25PySpark = df2PySpark.na.drop(how = 'any', subset = 'Salary')
df2PySpark.show(30)
df25PySpark.show(30)
```

```python
# Dropping or removing rows containing 'null' values with condition
# using 'subset = column name' condition
df26PySpark = df2PySpark.na.drop(how = 'any', subset = ['YearOfExperience','Salary'])
df2PySpark.show(30)
df26PySpark.show(30)
```

#### Handling missing values

```python
# Replacing all missing values of type 'String' in PySpark DataFrame
df27PySpark = df2PySpark.na.fill('Missing Value')
df2PySpark.show(30)
df27PySpark.show(30)
df27PySpark.printSchema()
```

```python
# Replacing all missing values if type 'int' in PySpark DataFrame
df28PySpark = df2PySpark.na.fill(0)
df2PySpark.show(30)
df28PySpark.show(30)
df28PySpark.printSchema()
```

```python
# Import Imputer class from PySpark
from pyspark.ml.feature import Imputer

# Create an instance of imputer for all missing values (numeric fields)
myImputer = Imputer(
                inputCols = ['YearOfExperience', 'Salary'],
                outputCols = ['{}_imputed'.format(col) for col in ['YearOfExperience', 'Salary']]
).setStrategy('mean')
```

```python
# Once myImputer is created, we can fit and transform it to DataFrame
myImputer.fit(df2PySpark).transform(df2PySpark).show(30)
```

```python
df2aPySpark = myImputer.fit(df2PySpark).transform(df2PySpark)
df2aPySpark.show(30)
df2aPySpark.printSchema()
```

```python
# Create an instance of imputer for all missing values (numeric fields)
myImputer = Imputer(
                inputCols = ['YearOfExperience', 'Salary'],
                outputCols = ['{}_imputed'.format(col) for col in ['YearOfExperience', 'Salary']]
).setStrategy('median')
```

```python
df2bPySpark = myImputer.fit(df2PySpark).transform(df2PySpark)
df2PySpark.show(30)
df2bPySpark.show(30)
df2bPySpark.printSchema()
```

#### Filtering operations

```python
# show employee who have less than 5 years of working experience 
df2PySpark.filter('YearOfExperience<=5').show()
```

```python
# Alternative
df2PySpark.filter(df2PySpark['YearOfExperience'] <= 5).show()
```

```python
# select employee who have less than 5 years of working experience. Show only the name and years of experience
df2PySpark.filter('YearOfExperience <= 5').select(['Name', 'YearOfExperience']).show()
```

```python
# We can keep the results as a new DataFrame
df31PySpark = df2PySpark.filter('YearOfExperience <= 5')
df31PySpark.show()
df31PySpark.printSchema()
```

```python
# Similarly, we can filter the rows from a dataframe based on certain condition, and suppress certain attribute.
df32PySpark = df2PySpark.filter('YearOfExperience <= 5').select(['Name', 'YearOfExperience'])
df32PySpark.show()
df32PySpark.printSchema()
```

```python
# Selecting rows with multiple conditions logically 'AND'
df2PySpark.filter((df2PySpark['YearOfExperience'] <= 5) & (df2PySpark['Salary'] >= 60000)).show()
```

```python
# Selecting rows with multiple conditions logically 'OR'
df2PySpark.filter((df2PySpark['YearOfExperience'] <= 5) | (df2PySpark['Salary'] >= 60000)).show()
```

```python
# Selecting rows with multiple conditions
df2PySpark.filter(~((df2PySpark['YearOfExperience'] <= 5) | (df2PySpark['Salary'] >= 60000))).show()
df2PySpark.show(50)
```

```python
# We can return the result of the filter operation as a class of rows (list) using the collect() method
df2PySpark.filter((df2PySpark['YearOfExperience'] <= 5) & (df2PySpark['Salary'] >= 60000)).collect()
```

```python
temp = df2PySpark.filter((df2PySpark['YearOfExperience'] <= 5) & (df2PySpark['Salary'] >= 60000)).collect()
print(temp)
```

# PySpark GrouBy and Aggregate Functions

```python
# Reading a new practice data set
df3PySpark = spark.read.csv('PracticeData3.csv', header = True, inferSchema = True)
df3PySpark.show(30, False)
df3PySpark.printSchema()
```

#### GroupBy operations

```python
# GroupBy
# What is groupby return?
df3PySpark.groupby('Project')
```

```python
# Find the total salary of all employee working in a specific project.
# Note if no parameter is specified in the sum() method, all numeric typed columns will be computed
df3PySpark.groupBy('Project').sum().show(50, False)
```

```python
# Find the total salary of all employee working in a specific project.
df3PySpark.groupBy('Project').sum('Salary').show(50, True)
```

```python
# Find the total salary of all employee working in a specific project.
df3PySpark.groupBy('Project').sum('YearOfExperience').show(50, truncate = False)
```

```python
# Find what is the mean (average) salary of employees working in each project
df3PySpark.groupBy('Project').avg('Salary').show(50,False)
```

```python
# Find what is the mean (average) salary of employees working in each project
df3PySpark.groupBy('Project').mean('Salary').show(50,False)
```

```python
# Count how many employees are working in each project
df3PySpark.groupBy('Project').count().show(50,False)
```

#### Aggregate operations

```python
# Aggregate the entire class (DataFrame without groups) and find the total salary
df3PySpark.agg({'Salary':'sum'}).show()
```

```python
# Aggregate the entire class (DataFrame without groups) and find the largest (highest) salary
df3PySpark.agg({'Salary':'max'}).show()
```

```python
# Aggregate the entire class (DataFrame without groups) and find the largest value
df3PySpark.agg({'Name':'max'}).show()
```

```python
# Aggregate the entire class (DataFrame without groups) and find the highest year of working experience
df3PySpark.agg({'YearOfExperience':'max'}).show()
```

```python
# Aggregate the entire class (DataFrame without groups) and find the lowest year of working experience
df3PySpark.agg({'YearOfExperience':'min'}).show()
```
